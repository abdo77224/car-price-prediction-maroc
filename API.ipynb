```python
import pandas as pd
import requests
import time

# Charger votre dataset
data = pd.read_csv('Avito_Cars_hssn.csv')

# Configurer l'API
API_URL = "https://api.api-ninjas.com/v1/cars"
API_KEY = "AavLY7kdj03TbRQvgJjFeA==2tDtJetDeCAJKeu9"  # Remplacez par votre clé API

# Fonction pour récupérer le type de carburant via l'API
def get_fuel_type(make, model):
    headers = {"X-Api-Key": API_KEY}
    params = {"make": make, "model": model}
    
    try:
        response = requests.get(API_URL, headers=headers, params=params)
        
        # Vérifiez si la requête a réussi
        if response.status_code == 200:
            json_data = response.json()
            # Vérifiez si les données existent
            if len(json_data) > 0:
                return json_data[0].get('fuel_type', 'Inconnu')  # Récupérer le type de carburant ou 'Inconnu'
            else:
                return "Inconnu"
        else:
            print(f"Erreur {response.status_code} pour {make} {model}")
            return "Inconnu"
    except Exception as e:
        print(f"Erreur lors de l'appel API : {e}")
        return "Inconnu"

# Appliquer la fonction sur chaque ligne du dataset
data['fuel'] = data.apply(lambda row: get_fuel_type(row['Marque'], row['Model']), axis=1)

# Sauvegarder le dataset mis à jour
data.to_csv('dataset_avec_carburant.csv', index=False)
print("Les types de carburant ont été ajoutés avec succès !")
```

```python
import pandas as pd
from tqdm import tqdm
import requests

# Charger votre dataset
data = pd.read_csv('Avito_Cars_hssn.csv')

# Définir l'URL de l'API et la clé API
API_URL = "https://api.api-ninjas.com/v1/cars"
API_KEY = "AavLY7kdj03TbRQvgJjFeA==2tDtJetDeCAJKeu9"  # Remplacez par votre clé API

# Cache pour éviter des appels répétitifs pour les mêmes marque/modèle
cache = {}

# Fonction pour interagir avec l'API et utiliser le cache
def get_fuel_type_with_cache(make, model):
    key = f"{make}_{model}"  # Clé unique pour marque + modèle
    if key in cache:  # Si déjà dans le cache, renvoyer le résultat
        return cache[key]
    try:
        headers = {"X-Api-Key": API_KEY}
        params = {"make": make, "model": model}
        response = requests.get(API_URL, headers=headers, params=params)
        if response.status_code == 200:
            json_data = response.json()
            fuel_type = json_data[0].get('fuel_type', 'Inconnu') if len(json_data) > 0 else 'Inconnu'
            cache[key] = fuel_type  # Ajouter au cache
            return fuel_type
        else:
            print(f"Erreur {response.status_code} pour {make} {model}")
            return "Inconnu"
    except Exception as e:
        print(f"Erreur API : {e}")
        return "Inconnu"

# Traiter tout le dataset avec suivi et sauvegardes régulières
tqdm.pandas()  # Activer l'intégration de tqdm avec pandas
data['fuel'] = data.progress_apply(lambda row: get_fuel_type_with_cache(row['Marque'], row['Model']), axis=1)

# Sauvegarder le dataset final
data.to_csv('dataset_avec_carburant.csv', index=False)
print("Traitement terminé, dataset final sauvegardé avec la colonne 'fuel'.")
```
